```python
class Model(nn.Module):
    """
        经过三层全连接处理
            - 参数共享机制
                - 这三层是一样的！
                    - 只有一套参数
    """
    def __init__(self, num_features=128, N=3):
        super().__init__()
        self.N = N
        self.linear = nn.Linear(in_features=num_features, out_features=num_features)

    def forward(self, x):
        # 处理 3 次
        for _ in range(self.N):
            x = self.linear(x)
        return x
```

```python
x = torch.randn(2, 128)
```

```python
model = Model(num_features=128, N=3)
```

```python
model
```

```python
model(x)
```

```python
for param in model.named_parameters():
    print(param)
```

```python
class Model(nn.Module):
    """
        经过三层全连接处理
            - 参数不共享
                - 这三层是不一样的！
                    - 有三套参数
    """
    def __init__(self, num_features=128, N=3):
        super().__init__()
        self.N = N
        self.linears = [nn.Linear(in_features=num_features, out_features=num_features) for _ in range(self.N)]

    def forward(self, x):
        # 处理 3 次
        for linear in self.linears:
            x = linear(x)
        return x
```

```python
class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.w = nn.Parameter(torch.ones(2, 5))
    def forward(self, x):
        return x @ self.w
```

```python
model = Model()
```
