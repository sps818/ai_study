### 3. 概率
- probability
    - 或然率，概率，频率，可能性，置信度，打分 ...
    - 变量有两种类型：
- 离散型变量：
    - 有限个状态
    - 状态之间无大小、程度等概念
    - 状态之间是严格对立的！！！
    - `概率的计算`：
        - `转换为频率`
        - `数个数`
        - `(y_pred == y_test).mean()`
- 连续型变量
    - 无限个数值
    - 数值之间是大小、程度的差异
    - 数值之间的内涵是一致的
    - 概率的计算：
        - PDF(Probability Density Function)
        - 通过对 **概率密度函数** 的积分来求解
        - **概率密度函数** 是 **概率** 的导数
        - **概率** 就是 **概率密度函数** 的积分
        - 计算步骤（理论数学）：
            - 先找 出概率密度函数来
            - 然后再对概率密度函数积分
            - 在任何一点的概率都是零
        - 计算步骤（工程数学）：
            - 把连续型变量的分布看作是高斯分布
            - 直接拿概率密度函数的值代替概率

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(start=-5, stop=5, num=100)
```

![normal.svg](normal.svg)

```python
# 上图中mu是均值、sigma是标准差
def normal(x, mu=0, sigma=1):
    """
    """
    return 1 / (2 * np.pi) ** 0.5 / sigma * np.exp(-(x - mu) ** 2 / 2 / sigma ** 2)    
```

```python
"""
数学知识点回顾：
正态分布这个曲线的面积等于1，也就是概率100%
mu就是对称轴，所以也可以叫location，他可以控制左右
标准差sigma越大，数据越分散，也就越扁平
"""
# 折线图
plt.plot(x, normal(x, mu=0, sigma=1), label="$\mu=0,\sigma =1$")  # 标准正态
plt.plot(x, normal(x, mu=1, sigma=1), label="$\mu=1,\sigma =1$")
plt.plot(x, normal(x, mu=0, sigma=2), label="$\mu=0,\sigma =2$")
plt.grid()
plt.legend()
```
