```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = plt.imread(fname="./boy.jpg")
```

```python
# 
plt.imshow(X=img)
```

```python
img_r = img[:, :, 0]
```

```python
plt.imshow(X=img_r, cmap="gray")
```

```python
"""
    
"""
N = 11
my_kernel1 = np.ones(shape=(N, N)) / N ** 2
img1 = cv2.filter2D(src=img_r, ddepth=-1, kernel=my_kernel1)
plt.imshow(X=img1, cmap="gray")
```

```python
"""
    
"""
my_kernel2 = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
img1 = cv2.filter2D(src=img_r, ddepth=-1, kernel=my_kernel2)
plt.imshow(X=img1, cmap="gray")
```

```python
"""
    
"""
kernel = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]]).T
img1 = cv2.filter2D(src=img_r, ddepth=-1, kernel=kernel)
plt.imshow(X=img1, cmap="gray")
```

```python
"""
    
"""
kernel = np.array([[1, 1, 1], [1, -8, 1], [1, 1, 1]]).T
img1 = cv2.filter2D(src=img_r, ddepth=-1, kernel=kernel)
plt.imshow(X=img1, cmap="gray")
```
