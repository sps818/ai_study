### 1. Matplotlib
- pip install matplotlib -U

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# 1. 读取图像
img = plt.imread(fname="girl.jpg")
```

```python
# [H, W, C] RGB
img.shape
```

```python
# 2. 图像显示
plt.imshow(X=img)
```

```python
# 处理1：读取 R 层
img_r = img[:, :, 0]
```

```python
img_r
```

```python
plt.imshow(X=img_r, cmap="gray")
```

```python
# 处理2：截取图像上半部分
```

```python
H, W, C = img.shape
```

```python
plt.imshow(X=img[:H//2, :, :])
```

```python
# 右侧半张图像
plt.imshow(X=img[:, W//2:, :])
```

```python
# 处理3：左下角 1/ 4
plt.imshow(X=img[H//2:, :W//2, :])
```

```python
# 亚采样
plt.imshow(X=img[::4, ::4, :])
```

```python
# 保存
plt.imsave(fname="girl1.jpg", arr=img[::4, ::4, :])
```
