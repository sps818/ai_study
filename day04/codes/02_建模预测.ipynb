### 2.1 KNN算法

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X=X_train, y=y_train)
```

```python
y_pred = knn.predict(X=X_test)
```

```python
acc = (y_pred == y_test).mean()
```

```python
acc
```

```python
# 初始化相关参数
best_n_neighbor = 0
best_acc = 0
best_model = None
# 简单遍历，寻找比较好的 N
for n_neighbor in range(3, 101, 1):
    knn = KNeighborsClassifier(n_neighbors=n_neighbor)
    knn.fit(X=X_train, y=y_train)
    y_pred = knn.predict(X=X_test)
    acc = (y_pred == y_test).mean()
    if acc > best_acc:
        best_acc = acc
        best_n_neighbor = n_neighbor
        best_model = knn
        print(f'找到一个更好的模型：n_neighbor = {best_n_neighbor}, acc = {best_acc}')
```

```python
# 最好邻居数
best_n_neighbor
```

```python
# 最好正确率
best_acc
```

```python
# 保存模型
joblib.dump(value=[best_n_neighbor, best_acc, best_model], filename="best_knn.lxh")
```

```python
# 加载模型
joblib.load(filename="best_knn.lxh")
```

### 3. 推理过程

```python
# 推理的时候，是从数据中抽取随机一条进行比如："787584108,Existing Customer,55,M,3,Unknown,Married,$80K - $120K,Blue,47,4,2,3,3436,2016,1420,0.901,1097,33,0.833,0.587"
```

```python
x = "787584108,55,M,3,Unknown,Married,$80K - $120K,Blue,47,4,2,3,3436,2016,1420,0.901,1097,33,0.833,0.587"
```

```python
x1 = "719808558,55,F,2,Graduate,Married,Less than $40K,Blue,43,2,4,3,1438.3,0,1438.3,0.707,886,27,0.421,0"
```

```python
def predict(x):
    """
        推理函数
    """
    import numpy as np
    print(f"原始输入：{x}")
    x = x.split(",")[1:]
    print(f"切分之后：{x}")
    temp = []
    # 1, age
    temp.append(float(x[0]))
    # 2, gender
    temp.append(gender_dict[x[1]])
    # 3, 
    temp.append(float(x[2]))
    # 4, 
    temp.append(education_dict[x[3]])

    # 5, 
    temp.append(marital_dict[x[4]])
    # 6，
    temp.append(income_dict[x[5]])

    # 7, 
    temp.append(card_dict[x[6]])

    # 8 - 19
    temp.extend([float(ele) for ele in x[7: ]])
    
    print(f"编码之后：{temp}")

    # 标准化
    x = np.array(temp)
    x = (x - mu) / sigma
    print(f"标准化后：{x}")

    # 进行模型推理
    y_pred = dtc.predict(X=[x])
    print(f"推理结果：{y_pred}")  

    # 推理结果是0，看不懂，所以进行结果解析
    final_result = idx2label[y_pred[0]]
    print(f"最终结果：{final_result}")
    
    return final_result 
```

```python
predict(x=x)
```

```python
predict(x=x1)
```
