```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# scores = np.random.randint(low=0, high=101, size=(30,))
scores = np.random.randint(low=0, high=101, size=30)
```

```python
scores
```

```python
plt.plot(scores)
plt.title(label="original data")
```

```python
# 中心化
scores1 = scores - scores.mean()
```

```python
scores1
```

```python
plt.plot(scores1)
```

```python
# 归一化
min_ = scores.min()
max_ = scores.max()
```

```python
scores2 = (scores - min_) / (max_ - min_)
```

```python
scores2
```

```python
plt.plot(scores2)
```

```python
# 标准化
mu = scores.mean()
sigma = scores.std()
```

```python
scores3 = (scores - mu) / sigma
```

```python
scores3
```

```python
plt.plot(scores3)
```
