假设函数：$ y = k_1x_1 + k_2x_2 + ... + k_nx_n + b $

怕你看懂，改成： $ y = w_1x_1 + w_2x_2 + ... + w_nx_n + b $ ， $w$就是权重， $b$叫偏置

### 1. 读取数据

```python
import pandas as pd
```

```python
data = pd.read_csv(filepath_or_buffer="./boston_house_prices.csv", skiprows=1)
```

```python
data
```

```python
data.info()
```

```python
X = data.drop(columns=["MEDV"]).to_numpy()
y = data["MEDV"].to_numpy()
```

### 2. 切分&标准化

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=0)
```

```python
X.shape
```

```python
y.shape
```

```python
mu = X_train.mean(axis=0)
sigma = X_train.std(axis=0)
```

```python
# 标准化
X_train1 = (X_train - mu) / sigma
X_test1 = (X_test - mu) / sigma
```

```python
# 查看结果，量纲已统一
print(f'X_train1:{X_train1}\n')
print(f'X_test1:{X_test1}')
```

### 3. 套用模型

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X=X_train1, y=y_train)
```

```python
# 上面说到了有 w，b
# 训练完的模型，这两个参数是固定下来了,有13个特征，所以有13个w，查看下权重：
lr.coef_
```

```python
# 查看下偏置b
lr.intercept_
```

```python
y_pred_lr = lr.predict(X=X_test1)
```

```python
y_pred_lr
```

```python
y_test
```

```python
# 由于是连续量，所以不存在准确率一说。所以衡量模型时直接做差
# 预测大了就是正数，预测小了就是负数，绝对值越小说明预测的越准
y_pred_lr - y_test
```

```python
# 这时有几个衡量指标：    平均绝对误差.即 MAE(Mean Absolute Error)
abs(y_pred_lr - y_test).mean()
```

```python
# 平均平方误差.即 MSE(Mean Squared Error)
# 它会把大的误差放的更大，小的误差缩的更小，以后更多的会用MSE
loss_lr = ((y_pred_lr - y_test) ** 2).mean()
print(f'KNN的平均平方误差：{loss_lr}')
```

```python
# 用KNN尝试下
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X=X_train1, y=y_train)
y_pred_knn = knn.predict(X=X_test1)
loss_knn = ((y_pred_knn - y_test) ** 2).mean()
print(f'KNN的平均平方误差：{loss_knn}')
```

```python
# 用决策树尝试下
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor()
dtr.fit(X=X_train1, y=y_train)
y_pred_dtr = dtr.predict(X=X_test1)
loss_dtr = ((y_pred_dtr - y_test) ** 2).mean()
print(f'DTR的平均平方误差：{loss_dtr}')
```

### 逻辑回归
- 二分类算法
- 在线性回归的基础上，增加了一个 sigmoid 激活函数
    - 线性回归的值域是 $(+\infty, -\infty)$ ，但是sigmoid激活函数是S型曲线，套上它之后，原本的线性回归的值域被压缩在了 $(0, 1)$
    - 所谓二分类算法，就是大于0.5就是第一类，小于0.5就是第二类  可以这么搞
- 优化：梯度下降法 SGD: Stoc**h**astic Gradient Descent
