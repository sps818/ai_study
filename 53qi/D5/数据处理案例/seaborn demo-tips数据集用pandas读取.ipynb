```python
import seaborn as sns
import pandas as pd
```

tips 数据集是一个经典的示例数据集，用于演示统计学方法和数据可视化技术。它包含了一个餐厅的服务员收集的关于客户的小费金额和其他信息的数据。该数据集中的变量包括：total_bill（账单总金额）、tip（小费金额）、sex（顾客性别）、smoker（是否吸烟）、day（就餐日期）、time（就餐时间）和 size（就餐人数）。  

这个数据集可以用于分析小费金额和其他变量之间的关系，比如性别、吸烟习惯、就餐日期和时间等。由于数据集比较简单易懂，所以它常常被用来进行数据可视化和探索性数据分析的演示。  

```python
# tips = sns.load_dataset("tips")
tips= pd.read_csv("tips.csv")
```

小提琴图和箱线图都是常用的用于可视化数据分布的图形，它们的主要区别在于：

形状：箱线图是基于五数概括（最小值、第一四分位数、中位数、第三四分位数、最大值）来描述数据分布，所以它通常呈现为一个包含盒子和两条线段的图形。而小提琴图则会呈现出类似小提琴形状的曲线，以展示数据的密度分布情况。

显示范围：箱线图只能显示数据的五个统计量（最小值、第一四分位数、中位数、第三四分位数、最大值），而忽略了数据的具体分布情况。小提琴图则可以显示数据分布的整个范围，以及具体的密度分布情况。

可读性：小提琴图通常可以更好地展示数据分布的整体特征和密度，但它的可读性可能会受到分布形状的影响。相比之下，箱线图通常更容易被理解和解释，特别是对于非专业人士来说。

综上所述，小提琴图和箱线图各有其优缺点，在具体情况下需要根据数据类型和分析目的来选择使用哪种图形。

```python
#箱图
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips, order=None, hue_order=None, palette=None)
```

```python
#小提琴图
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, order=None, hue_order=None, palette=None, bw='scott')
```

```python
#点条图
sns.stripplot(x='day', y='total_bill', hue=None, data=tips, order=None, hue_order=None, palette=None,jitter=True, dodge=False)
```

```python
#直方图
sns.histplot(data=tips, x='total_bill', y=None, hue='smoker',weights=None, stat='count',bins='auto', binwidth=None, binrange=None, discrete=None, multiple='layer',kde=False)
```

```python
#条图
sns.countplot( x='sex', y=None, hue='smoker', data=tips, order=None, hue_order=None, orient=None, color=None, palette=None)
```

```python
#曲线图
sns.kdeplot(data=tips,x='total_bill',hue='time',shade=True,bw_method=0.1,multiple='layer')
```

```python
#分布图
sns.displot(data=tips, x='tip', hue='sex', row=None, col=None, kind='hist',kde=True)
```

在条形图（bar plot）中，垂直的黑线通常被称为误差线（error bar），用于表示在每个类别中数值变量的方差或标准差范围。误差线可以帮助观察者更好地理解数据的不确定性，以及不同类别之间数量或比率差异的显著性。

误差线的长度通常代表方差或标准差的大小

```python
#条形图
sns.barplot( x='day', y='tip', hue='sex', data=tips, order=None, hue_order=None,ci=95)
```

```python
#散点图
sns.scatterplot( x="total_bill", y="tip", hue="time", style="time", size=None, data=tips, palette=None)
```

```python
#折线图
sns.lineplot(x='size', y='total_bill', hue='smoker', size=None, style=None, data=tips, palette=None, markers=None)
```

```python
#关系图
sns.relplot(
    data=tips, x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
#离散图
sns.catplot(x='day', y='total_bill', hue='smoker', data=tips, row='time', col='sex', kind='strip')
```

```python
#回归图
sns.regplot(data=tips, x='total_bill', y='tip', order=2, ci=50, marker='o', color=None)
```

```python
#多变量图
sns.pairplot(data=tips, hue=None, hue_order=None, palette=None, kind='scatter', diag_kind='auto', markers=None)
```
