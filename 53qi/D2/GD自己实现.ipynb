### 一维梯度下降

```python
import numpy as np
from matplotlib import pyplot as plt

def fn(x):
    """
        原函数
    """
    return (x - 3) ** 2

def dfn(x):
    """
        导函数
    """
    return 2 * (x - 3)

def sgd(starting_point=0, learning_rate=0.1, num_iterations=100):
    """
        Stochastic Gradient Descent  梯度下降函数
    """
    x = starting_point
    history = [x]  # 记录每一步的 x 值  
    
    for iteration in range(num_iterations):
        # 梯度下降
        x = x - learning_rate * dfn(x)
        history.append(x)
        if iteration % 10 == 0:
            print(f'优化 {iteration+1} 步后，x的值是：{x}')
    return x, history


# 随机的开始
starting_point = np.random.randint(low=-3, high=9, size=1)
# 循环的步数
num_iterations = 70
# 学习率
learning_rate = 0.1
# 用梯度下降法进行 优化
print(f'x的初始值：{starting_point}')
x, history = sgd(starting_point=starting_point, num_iterations=num_iterations)   
print(f'x的最终值：{x}') 

# 绘制过程
x_vals = np.linspace(-2, 8, 400)
y_vals = fn(x_vals)

plt.plot(x_vals, y_vals, label=r'$fn(x) = (x - 3)^2$', color='blue')
plt.scatter(history, [fn(x) for x in history], color='red', label='gradient descent', zorder=5)
plt.title('gradient descent 1D')
plt.xlabel('x')
plt.ylabel('fn(x)')
plt.legend()
plt.grid(True)
plt.show()
```

### 二维梯度下降

```python
import numpy as np
from matplotlib import pyplot as plt

def fn(x, y):
    """
        原函数
    """
    return (x - 3) ** 2 + (y - 2) ** 2

def dfn(x, y):
    """
        导函数
    """
    grad_x = 2 * (x - 3)
    grad_y = 2 * (y - 2)
    return np.array([grad_x, grad_y])

def sgd(starting_point=0, learning_rate=0.1, num_iterations=100):
    """
        Stochastic Gradient Descent  梯度下降函数
    """
    point = starting_point
    history = [point]  # 记录每一步的 x,y 值  
    
    for iteration in range(num_iterations):
        # 梯度下降
        point = point - learning_rate * dfn(point[0], point[1])
        history.append(point)
        if iteration % 10 == 0:
            print(f'优化 {iteration+1} 步后，x,y的值是：{point[0]},{point[1]}')
    return point, history


# 随机的开始
starting_point = np.random.randint(low=-3, high=9, size=2)
# 循环的步数
num_iterations = 70
# 学习率
learning_rate = 0.1
# 用梯度下降法进行 优化
print(f'x,y的初始值：{starting_point}')
point, history = sgd(starting_point=starting_point, num_iterations=num_iterations)   
print(f'x,y的最终值：{point}') 

# 绘制过程
x_vals = np.linspace(-2, 8, 400)
y_vals = np.linspace(-2, 6, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = fn(X, Y)

# 绘制目标函数的等高线图
plt.contour(X, Y, Z, levels=20, cmap='Blues')

# 绘制梯度下降的过程
history = np.array(history)
plt.plot(history[:, 0], history[:, 1], marker='o', color='red', label='gradient descent roat')

# 绘制最优解点
plt.scatter(point[0], point[1], color='green', s=100, label='best (3, 2)', zorder=5)

# 设置图形属性
plt.title('gradient descent 2D')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()  
```
