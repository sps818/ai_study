## 一维梯度下降

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义目标函数 f(x) = (x - 3)^2 和它的导数（梯度）
def f(x):
    return (x - 3) ** 2

def df(x):
    return 2 * (x - 3)

# 梯度下降函数
def gradient_descent(starting_point, learning_rate, num_iterations):
    x = starting_point
    history = [x]  # 记录每一步的 x 值
    
    for _ in range(num_iterations):
        grad = df(x)  # 计算当前点的梯度
        x = x - learning_rate * grad  # 更新 x 的值
        history.append(x)
    
    return x, history

# 初始设置
starting_point = 0  # 从 x=0 开始
learning_rate = 0.1  # 学习率
num_iterations = 50  # 迭代次数

# 执行梯度下降
optimal_x, history = gradient_descent(starting_point, learning_rate, num_iterations)

# 打印结果
print(f"最优解: x = {optimal_x}")

# 绘制过程
x_vals = np.linspace(-1, 7, 400)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label=r'$f(x) = (x - 3)^2$', color='blue')
plt.scatter(history, [f(x) for x in history], color='red', label='gradient descent', zorder=5)
plt.title('gradient descent 2D')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## 二维梯度下降

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义目标函数 f(x, y) = (x - 3)^2 + (y - 2)^2 和它的梯度
def f(x, y):
    return (x - 3) ** 2 + (y - 2) ** 2

def grad_f(x, y):
    grad_x = 2 * (x - 3)  # 对 x 的梯度
    grad_y = 2 * (y - 2)  # 对 y 的梯度
    return np.array([grad_x, grad_y])

# 梯度下降函数
def gradient_descent(starting_point, learning_rate, num_iterations):
    point = starting_point
    history = [point]  # 记录每一步的 (x, y) 值
    
    for _ in range(num_iterations):
        grad = grad_f(point[0], point[1])  # 计算当前点的梯度
        point = point - learning_rate * grad  # 更新 (x, y) 的值
        history.append(point)
    
    return point, history

# 初始设置
starting_point = np.array([0, 0])  # 从 (0, 0) 开始
learning_rate = 0.1  # 学习率
num_iterations = 50  # 迭代次数

# 执行梯度下降
optimal_point, history = gradient_descent(starting_point, learning_rate, num_iterations)

# 打印结果
print(f"最优解: (x, y) = {optimal_point}")

# 绘制过程
x_vals = np.linspace(-1, 7, 400)
y_vals = np.linspace(-1, 5, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

# 绘制目标函数的等高线图
plt.contour(X, Y, Z, levels=20, cmap='Blues')

# 绘制梯度下降的过程
history = np.array(history)
plt.plot(history[:, 0], history[:, 1], marker='o', color='red', label='gradient descent roat')

# 绘制最优解点
plt.scatter(optimal_point[0], optimal_point[1], color='green', s=100, label='best (3, 2)', zorder=5)

# 设置图形属性
plt.title('gradient descent 2D')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
