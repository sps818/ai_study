```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
# 生成示例数据
features, label = make_blobs(n_samples=500, n_features=2, centers=4, cluster_std=1.5, random_state=42)
plt.scatter(x=features[:, 0], y=features[:, 1], c=label)
```

```python
# 尝试不同的k值
range_n_clusters = range(2, 11)
silhouette_scores = []

for n_clusters in range_n_clusters:
    # 进行聚类
    clusterer = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = clusterer.fit_predict(features)
    
    # 计算轮廓系数
    silhouette_avg = silhouette_score(features, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For n_clusters = {n_clusters}, the average silhouette_score is: {silhouette_avg:.4f}")

# 绘制轮廓系数随k值变化的曲线
plt.plot(range_n_clusters, silhouette_scores, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different k Values')
plt.grid(True)
plt.show()
# 轮廓系数法验证：从下面输出找到最接近1的，它就是分类最好的。比如下面输出分成3个类可能最好。
```
