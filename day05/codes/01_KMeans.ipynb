```python
from sklearn.datasets import make_blobs  # load相关的都是真实数据，make相关的假数据生成的办法，blobs是聚类相关的
from matplotlib import pyplot as plt
```

```python
"""
    参数说明：
    n_samples:生成多少个样本
    n_features:生成多少个特征
    centers:想要几个类别
    cluster_std: 标准差。数值越大越分散，数据越小越聚合
"""
X, y = make_blobs(n_samples=1000, 
                  n_features=2, 
                  centers=4,
                  cluster_std=1, 
                  random_state=0)
```

```python
plt.scatter(x=X[:, 0], y=X[:, 1], c=y)
```

```python
"""
    分类 4 类
"""
```

```python
from sklearn.cluster import KMeans
```

```python
"""
    不管是有监督还是无监督学习
    sklean api 高度一致的！！
"""
```

```python
# 实例化对象，分成4类
km = KMeans(n_clusters=4, n_init='auto')
```

```python
# 训练
km.fit(X=X, y=y)
```

```python
km.cluster_centers_
```

```python
plt.scatter(x=X[:, 0], y=X[:, 1], c=y)
# 再把找到的4个中心画到一起
"""
    参数说明：
    c:标颜色
    marker:画出来的点的形状
    s:marker的大小
"""
plt.scatter(x=km.cluster_centers_[:, 0], y=km.cluster_centers_[:, 1], c='red', marker='*', s=100)
```
