```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def fn(x):
    """
        定义函数
    """
    return x ** 2
```

```python
x = np.linspace(start=-5, stop=5, num=100)
```

```python
"""
    绘制图形
"""
plt.plot(x, fn(x))
plt.grid()
```

```python
def dfn(x):
    """
        求导函数
    """
    return 2 * x
```

```python
# 走一个固定的步数
steps = 1000
# 把步子压得小一点
learning_rate = 1e-2
```

```python
# 1，随机的开始
x = np.random.randint(low=-1000, high=1001, size=(1,))
print(f"x的初始值: {x}")
# 2，迭代优化
for step in range(steps):
    # 梯度下降法
    x = x - learning_rate * dfn(x)
    print(f"优化{step+1}步后, x的值: {x}")
print(f"x的最终值: {x}")
```
